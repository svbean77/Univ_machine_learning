```python
#기계학습응용 12주차 과제
import numpy as np
from keras.layers import SimpleRNN, Dense
from keras.models import Sequential
```

```python
X = []
Y = []
for i in range(1, 5):
    lst = list(range(i, i + 3))
    X.append(list(map(lambda x: x/10, lst)))
    Y.append(list(map(lambda y: (y + 1)/10, lst)))
    
train_X = np.array(X).reshape(-1, 3, 1)
train_Y = np.array(Y).reshape(-1, 3, 1)
test_X = np.array([[[0.5], [0.6], [0.7]],
                  [[0.6], [0.7], [0.8]]])
print(test_X.shape)
train_X.shape
```

```python
model = Sequential()
model.add(SimpleRNN(100, return_sequences = True, input_shape = (3, 1)))
model.add(Dense(1))
model.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])
model.fit(train_X, train_Y, epochs = 200, verbose = 2)
```

```python
print("vanilla 모델에 대한 성능 평가\n")
print(model.summary())
pred_Y = model.predict(test_X)
pred_Y
```

```python
model = Sequential()
model.add(SimpleRNN(100, return_sequences = True, input_shape = (3, 1)))
model.add(SimpleRNN(100, return_sequences = True))
model.add(Dense(1))
model.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])
model.fit(train_X, train_Y, epochs = 200, verbose = 2)
```

```python
print("stacked 모델에 대한 성능 평가\n")
print(model.summary())
pred_Y = model.predict(test_X)
pred_Y
```
