```python
#기계학습응용 3주차 과제
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = np.loadtxt('weightHeight.csv', delimiter=",")
cond = np.array([data[i][0] == 0 for i in range(10000)])  #남자를 true, 여자를 false로 조건 생성
#성별별 몸무게, 키를 n행 1열로 자름(남자는 true, 여자는 false에 속하는 구분이 들어가도록)
man_weight, man_height = data[cond, np.newaxis, 1], data[cond, np.newaxis, 2]
woman_weight, woman_height = data[~cond, np.newaxis, 1], data[~cond, np.newaxis, 2]

print('남자 몸무게와 키의 상관관계')
manW_train = man_weight[:-100]  #훈련을 진행할 뒤에서 100개를 제외한 남자 몸무게 샘플
manW_test = man_weight[-100:]  #테스트를 진행할 뒤에서 100개의 남자 몸무게 샘플
manH_train = man_height[:-100]  #훈련을 진행할 뒤에서 100개를 제외한 남자 키 샘플
manH_test = man_height[-100:]  #테스트를 진행할 뒤에서 100개의 남자 키 샘플

regM = LinearRegression()  #모델 설정
regM.fit(manW_train, manH_train)  #훈련시킬 샘플 데이터를 넣음
manH_pred = regM.predict(manW_test)  #테스트를 진행할 남자의 몸무게 샘플로 키를 예측
plt.scatter(manW_test, manH_test, color = 'skyblue')
plt.plot(manW_test, manH_pred, color = 'black', linewidth = 1)
plt.show()

print('여자 몸무게와 키의 상관관계')
womanW_train = woman_weight[:-100]  #훈련을 진행할 뒤에서 100개를 제외한 여자 몸무게 샘플
womanW_test = woman_weight[-100:]  #테스트를 진행할 뒤에서 100개의 여자 몸무게 샘플
womanH_train = woman_height[:-100]  #훈련을 진행할 뒤에서 100개를 제외한 여자 키 샘플
womanH_test = woman_height[-100:]  #테스트를 진행할 뒤에서 100개의 여자 키 샘플

regW = LinearRegression()  #모델 설정
regW.fit(womanW_train, womanH_train)  #훈련시킬 샘플 데이터를 넣음
womanH_pred = regW.predict(womanW_test)  #테스트를 진행할 여자의 몸무게 샘플로 키를 예측
plt.scatter(womanW_test, womanH_test, color = 'pink')
plt.plot(womanW_test, womanH_pred, color = 'black', linewidth = 1)
plt.show()
```
